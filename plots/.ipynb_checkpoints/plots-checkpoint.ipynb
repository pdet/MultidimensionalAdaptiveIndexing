```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import sqlite3
import pandas as pd
import json

# pd.set_option('display.max_rows',5000)
```

```python
conn = sqlite3.connect("results")
df = pd.read_sql_query("SELECT * FROM RESULTS;", conn);

with open('config.json') as json_file:
    config = json.load(json_file)
```

```python
def average_each_query(list_of_values):
    repetitions = int(config['repetitions'])
    final = []
    length = len(list_of_values)/repetitions
    length = int(length)
        
    for i in range(length):
        final.append(0.0)
        for r in range(repetitions):
            final[i] += list_of_values[i + r * length]
        final[i] = final[i]/float(repetitions)
    return final
    

def cum_sum_plot(df):
    fig = make_subplots(rows=1, cols=1)
    algorithms = df['NAME'].unique()
    sums = {}
    for alg in algorithms:
        temp_df = df.loc[df['NAME'] == alg]
        temp_df['TOTAL_TIME']  = temp_df['INITIALIZATION_TIME']
        temp_df['TOTAL_TIME'] += temp_df['ADAPTATION_TIME']
        temp_df['TOTAL_TIME'] += temp_df['QUERY_TIME']
        fig.add_scatter(
            y=average_each_query(
                list(temp_df['TOTAL_TIME'].cumsum())
            ),
            name=alg
        )
    fig.update_xaxes(title_text="Query Number")
    fig.update_yaxes(title_text="Time (seconds)")
    fig.update_layout(title_text=f"""Cumulative Sum
                     {config['number_of_attributes']} column(s)
                     {config['number_of_tuples']} tuples
                     {config['selectivity']} selectivity""")
    fig.show()
```

```python
def per_query_plot(df_t):
    fig = make_subplots(rows=1, cols=1)
    algorithms = df['NAME'].unique()
    sums = {}
    for alg in algorithms:
        temp_df = df.loc[df['NAME'] == alg]
        temp_df['TOTAL_TIME']  = temp_df['INITIALIZATION_TIME']
        temp_df['TOTAL_TIME'] += temp_df['ADAPTATION_TIME']
        temp_df['TOTAL_TIME'] += temp_df['QUERY_TIME']
        fig.add_scatter(
            y=average_each_query(
                list(temp_df['TOTAL_TIME'])
            ),
            name=alg
        )
    fig.update_xaxes(title_text="Query Number")
    fig.update_yaxes(title_text="Time (seconds)")
    fig.update_layout(title_text=f"""Per Query
                     {config['number_of_attributes']} column(s)
                     {config['number_of_tuples']} tuples
                     {config['selectivity']} selectivity""")
    fig.show()
```

```python
cum_sum_plot(df)
```

```python
per_query_plot(df)
```
